## Numpy基本练习

```python
import numpy as np
import time
```

```python
A = np.random.rand(100,100)
B = np.random.rand(100,100)
Alist = A.tolist();
Blist = B.tolist();
```

* 请在下面用三重循环写出Alist和Blist矩阵相乘

```python
time_start=time.time()
# 请在下面写代码
Clist = []
for i in range(len(Alist)):
    Crow = []
    for j in range(len(Blist[0])):
        c = 0
        for k in range(len(Alist[i])):
            c += Alist[i][k] * Blist[k][j]
        Crow.append(c)
    Clist.append(Crow)
time_end = time.time()
print ("Time elapse:",(time_end - time_start))
```

## 使用Numpy处理图像数据

```python
from PIL import Image
from numpy import asarray
# load the image
image = Image.open('koala.jpeg')
# convert image to numpy array
data = asarray(image)
print(type(data))
# summarize shape
print(data.shape)

# create Pillow image
image2 = Image.fromarray(data)
print(type(image2))

# summarize image details
print(image2.mode)
print(image2.size)
```

```python
# load and display an image with Matplotlib
from matplotlib import image
from matplotlib import pyplot
# load image as pixel array
image = image.imread('koala.jpeg')
# summarize shape of the pixel array
print(image.dtype)
print(image.shape)
# display the array of pixels as an image
pyplot.imshow(image)
pyplot.show()
```

```python
im = np.array(Image.open('koala.jpeg').convert('L'))
gr_im= Image.fromarray(im).save('gr_koala.png')
```

```python
load_img_rz = np.array(Image.open('koala.jpeg').resize((200,200)))
Image.fromarray(load_img_rz).save('r_koala.jpeg')
print("After resizing:",load_img_rz.shape)
```

```python
im = np.array(Image.open('koala.jpeg'))

print("Before trimming:",im.shape)

im_trim = im[128:384, 128:384]
print("After trimming:",im_trim.shape)

Image.fromarray(im_trim).save('trim_koala.jpeg')
```
